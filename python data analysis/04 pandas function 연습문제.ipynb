#### 연습문제 3
- 타이타닉호 승객을 사망자와 생존자 그룹으로 나누고(alive)
- 각 그룹에 대해 미성년자, 청년, 중년, 장년, 노년 승객의 비율을 구하시오.
    - bins=[1,18,35,50,65,99]
    - labels=['미성년자','청년','중년','장년','노년']
    - 각 그룹별 비유의 전체 합은 1이 되어야 한다.

```python
alive_yes = titanic[titanic['alive']== 'yes'] ;alive_yes
alive_no = titanic[titanic['alive']== 'no'] ; alive_no
```

```python
bins=[1,18,35,50,65,99]
labels=['미성년자','청년','중년','장년','노년']

alive_yes_ctgs = pd.cut(alive_yes['age'], bins, labels = labels)

alive_no_ctgs = pd.cut(alive_no['age'], bins, labels = labels)
```

```python
alive_yes_ctgs.value_counts(normalize=True) * 100
alive_no_ctgs.value_counts(normalize=True) * 100
```
